```python
#Cleanin age column
def avgaccdtoclass(col):
    Age = col[0]
    Pclass= col[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 28
        else:
            return 24
    else:
        return Age
```

```python
train['Age']  = train[['Age','Pclass']].apply(avgaccdtoclass,axis=1)
#remove cabin and removena embarked
train.drop('Cabin',axis=1,inplace=True)
train.dropna(inplace=True)

#check missing data 
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
train.info()
```

```python
#numerical data required only for algorithm to run
sex = pd.get_dummies(train['Sex'],drop_first=True)
emb = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train = pd.concat([train,sex,emb],axis=1)
```

```python
train.head()
```

```python
train.drop(['Sex','Embarked','Name','Ticket','PassengerId'],axis=1,inplace=True)
```

```python
train.head()
```

```python
train.head()
```

```python
x = train.drop('Survived',axis=1)
y=train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mdl = LogisticRegression()
```

```python
mdl.fit(X_train,y_train)
```

```python
pred = mdl.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```
